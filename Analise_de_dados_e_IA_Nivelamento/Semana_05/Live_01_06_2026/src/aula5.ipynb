```python
modelo =['uno','gol','polo','kombi']
preco = [20000,30000,60000,15000]

for i in range(len(modelo)):
    print(f"O carro  tem preço{modelo[i]} tem preço:  RS: {preco[i]}")
```

```python
escolha = input("Diga o modelo do carro:")
for i in range(len(modelo)):
    if modelo[i] == escolha:
        print(f"O carro {escolha} tem preço {preco[i]}")
```

```python
carros = {
    'modelo':['uno','gol','polo','kombi'],
    'preço':[20000,30000,60000,15000],
    'cor': ['Branco','preto','branco','azul'],
    'potencia':[80,10,120,100],
    'ano':[1998,2005,2010,2015]
}
```

```python
import pandas as pd

carros = pd.DataFrame(carros)
carros.to_excel("carros.xlsx")
```

```python
dados = pd.read_excel("carros.xlsx")
dados
```

```python
dados[['modelo','preco']]
```

```python
preco_medio = dados['preco'].mean()
print(f"O preço médio dos carros na tabela é R$:{preco_medio}")
```

```python
dados[dados['preco'] > 30000]
```

```python
dados.sort_values(['preco'],ascending=False)
```

```python
carros_volks  = dados[dados['marca'] == 'Volkswagen']
print(f"O preço médio dos caros da volks é R$:{carros_volks['preco'].mean()}")
```

```python
for marca in dados['marca'].unique():
    carros_marca = dados[dados['marca'] == marca]
    print(f"O preço médio dos carros da {marca} é R$: {carros_marca['preco'].mean()}")
```

```python
automaticos = dados[dados['cambio'] =='automático']
print(f"O preço médio dos carros automáticos é R$:{automaticos['preco'].mean()}")
```

```python
precos = 0 
for preco in automaticos['preco']:
    precos += preco
    print(precos)
preco_medio =len(automaticos)

print(preco_medio)

print(f"O preço médio dos carros automáticos é R${automaticos['preco'].mean()}")
```
